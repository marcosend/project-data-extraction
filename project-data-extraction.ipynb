### Data extraction

Primero necesitamos importar las librerías que se necesitarán

```python
import pandas as pd
import requests
import json
from pandas.io.json import json_normalize
```

La data del API estaba muy bien organizada y fue relativamente sencillo descargar la data

```python
df = pd.DataFrame()
for i in range(668): #Dado que no había un dato de número de páginas totales era difícil poner este número dinámico
    url=f'https://api.magicthegathering.io/v1/cards?page={i+1}'
    respuesta = requests.get(url)
    resultado = respuesta.json()
    flattened_data = pd.json_normalize(resultado)
    flattened_data1 = pd.json_normalize(flattened_data.cards[0])
    flattened_data1 = flattened_data1.set_index('id', drop = False)
    df = df.append(flattened_data1, ignore_index = True)

df
```
